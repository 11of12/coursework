# Batting Average Example

```python
import numpy as np
team_names = ["Houston Astros", "Cininnati Reds", "Atlanta Braves", "Detroit Tigers", "New York Mets", "Seattle Mariners"]
x_data = [
    [0.267],
    [0.249],
    [0.244],
    [0.242],
    [0.238],
    [0.226],    
]
y_data = [95, 83, 88, 77, 77, 90]
```

```python
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt

N = 1000
x_star = np.linspace(0.2, 0.3, N)
x_star = [[x] for x in x_star]

neigh = KNeighborsRegressor(n_neighbors=1)
neigh.fit(x_data, y_data)

y_star = neigh.predict(x_star)

x_sorted = np.sort(np.array(x_data)[:, 0])
boundary_lines = x_sorted[0:-1] + np.diff(x_sorted) / 2
boundary_lines = np.delete(boundary_lines, 1)

for line in boundary_lines:
    plt.axvline(line, linestyle="--", color="r")


plt.plot(x_star, y_star)
plt.xlim([0.22, 0.28])
plt.xlabel("Team Batting Average")
plt.ylabel("Number of wins in 2021 regular season")
plt.title(f"Decision regions for k = 1")

print(neigh.predict([[0.248]]))

```

```python
neigh = KNeighborsRegressor(n_neighbors=3)
neigh.fit(x_data, y_data)

y_star = neigh.predict(x_star)

boundary_lines = [0.235, 0.2435, 0.2545]
for line in boundary_lines:
    plt.axvline(line, linestyle="--", color="r")

plt.plot(x_star, y_star)
plt.xlim([0.22, 0.28])
plt.xlabel("Team Batting Average")
plt.ylabel("Number of wins in 2021 regular season")
plt.title(f"Decision regions for k = 3")
print(neigh.predict([[0.248]]))
```

# World Happiness Report

```python
x_data = [
    [0.5192, 0.9124],
    [0.2321, 0.8034],
    [0.6176, 0.6093],
    [0.3057, 0.6314],
    [0.4338, 0.7171],
    [0.6335, 0.9534],
    [0.5588, 0.7106],
    [0.4046, 0.8157],
]

RED = 0
BLUE = 1

y_data = [RED, BLUE, RED, RED, BLUE, BLUE, BLUE, RED]

```

```python
from sklearn.neighbors import KNeighborsClassifier
# https://www.geeksforgeeks.org/machine-learning/how-to-visualize-knn-in-python/
from mlxtend.plotting import plot_decision_regions

neigh = KNeighborsClassifier(n_neighbors=1)
neigh.fit(x_data, y_data)

plot_decision_regions(np.array(x_data), np.array(y_data), clf=neigh, colors='red,blue', zoom_factor=2.8, legend=0)
plt.xlim([0, 0.9])
plt.ylim([0.4, 1.1])
plt.xlabel("confidence in nat. govt.")
plt.ylabel("freedom to make life choices")
plt.title("Generosity - Red High, Blue Low (k = 1)")
neigh.predict([[0.5, 0.75]])
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_data, y_data)

plot_decision_regions(np.array(x_data), np.array(y_data), clf=neigh, colors='red,blue', zoom_factor=2.8, legend=0)
plt.xlim([0, 0.9])
plt.ylim([0.4, 1.1])
plt.xlabel("confidence in nat. govt.")
plt.ylabel("freedom to make life choices")
plt.title("Generosity - Red High, Blue Low (k = 3)")
neigh.predict([[0.5, 0.75]])
```
