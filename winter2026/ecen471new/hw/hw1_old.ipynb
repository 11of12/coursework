```python
from matplotlib import pyplot as plt
import numpy as np

def _calculate_k_nearest_neighbors(
    x: np.ndarray,
    x_data: np.ndarray,
    k: int = 1,
):
    x = np.atleast_1d(x)
    num_data_points = x_data.shape[0]
    kNN = np.zeros((num_data_points,k))
    for i in range(num_data_points):
        current_x = x_data[i, :]
        _diff = np.vstack([current_x]*num_data_points) - x_data
        euclidean_distances = [np.sqrt(np.dot(_diff[_i, :], _diff[_i, :])) for _i in range(num_data_points)]
        kNN[i, :] = np.argsort(euclidean_distances)[0:k]
    
    return kNN


# def _calculate_decision_region_boundaries(
#     x: np.ndarray,
#     kNN: np.ndarray
# ):

#     decision_boundaries = None
#     return decision_boundaries


def _regression_decision_regions(
    x: np.ndarray,
    x_data: np.ndarray,
    y_data: np.ndarray,
    k: int,
):
    kNN = _calculate_k_nearest_neighbors(x, x_data, k)
    # decision_boundaries = _calculate_decision_region_boundaries(x, kNN)
    pass

def _classification_decision_regions(
    x: np.ndarray,
    y: list,
    k: int
):
    pass


```
