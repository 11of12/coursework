```python
import numpy as np
x_data = np.array([
    [86.5, 10.1],
    [82.7, 13.2],
    [70.9, 17.1],
    [75.0, 11.7],
    [87.7, 12.4],
    [68.9, 30.7],
    [72.0, 25.9],
    [79.6,  8.1],
    [71.1, 20.4],
])

y_data = np.array([
    40.7,
    39.4,
    26.4,
    22.2,
    36.5,
    0.0,
    30.0,
    41.3,
    18.6,
])

free_throw_percentage = [[_data] for _data in x_data[:, 0]]
print(free_throw_percentage)
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt

regressor1 = DecisionTreeRegressor(criterion="squared_error", min_samples_split=4)
regressor1.fit(free_throw_percentage, y_data)

fig = plt.figure(figsize=(25, 20))
_ = plot_tree(regressor1, feature_names=["Free Throw %"])
```

```python
import matplotlib.pyplot as plt
import numpy as np

@np.vectorize
def calculate_z(a, b):
    return regressor2.predict([[a, b]])[0]

N = 100
X1, X2 = np.meshgrid(np.linspace(60, 90, N), np.linspace(0, 40))
Z = calculate_z(X1, X2)

plt.pcolormesh(X1, X2, Z)
plt.scatter(x_data[:, 0], x_data[:,1])
player_names = ["Grayson Allen", "Harrison Barnes", "Josh Giddey", "Elijah Hughes", "DeMar DeRozan", "Rudy Gobert", "Giannis Atetokounmpo", "Mike Conley", "Anthony Davis"]
for i, txt in enumerate(player_names):
    plt.annotate(txt, (x_data[i, 0], x_data[i, 1]))

plt.xlabel("Free Throw Percentage")
plt.ylabel("Defensive Rebound Percentage")
```

```python
import matplotlib.pyplot as plt
from matplotlib.colorizer import Colorizer
import numpy as np

x_data = np.array([
    [10.43, 5.72],
    [5.71, 4.14],
    [9.18, 6.18],
    [8.82, 5.15],
    [5.23, 2.47],
    [5.66, 3.14],
    [6.74, 2.73],
    [5.28, 2.91],
])

percentage_the = np.array([[_data] for _data in x_data[:, 0]])

y_data = ["Elder Bednar"]*4 + ["President Nelson"]*4
y_numeric = [ 0 if y =="Elder Bednar" else 1 if y=="President Nelson" else None for y in y_data]
```

```python
from sklearn.tree import DecisionTreeClassifier
from mlxtend.plotting import plot_decision_regions

classifier1 = DecisionTreeClassifier(criterion="entropy", min_samples_split=4)
classifier1.fit(percentage_the, y_data)

fig = plt.figure(figsize=(25, 20))
_ = plot_tree(classifier1, fontsize=40, feature_names=["% 'the'"], class_names=['Elder Bednar', "President Nelson"])
```

```python
classifier2 = DecisionTreeClassifier(criterion="entropy", min_samples_split=4)
classifier2.fit(x_data, y_data)

fig = plt.figure(figsize=(25, 20))
_ = plot_tree(classifier2, fontsize=40, feature_names=["% 'the'", "% 'of'"], class_names=['Elder Bednar', "President Nelson"])
```

```python
@np.vectorize
def calculate_z(x, y):
    author = classifier2.predict([[x, y]])[0]
    if author == "President Nelson":
        return 1.0
    elif author == "Elder Bednar":
        return 0.0

N = 100
X1, X2 = np.meshgrid(np.linspace(3.0, 12.0, N), np.linspace(2.0, 7.0))

Z = calculate_z(X1, X2)


plt.pcolormesh(X1, X2, Z)

plt.scatter(x_data[:4, 0], x_data[:4,1], label="Elder Bednar")
plt.scatter(x_data[4:, 0], x_data[4:,1], label="President Nelson")
plt.xlabel("% words = 'the'")
plt.ylabel("% words = 'of'")
plt.title("Decision tree considering x1 and x2")
plt.legend()
```
