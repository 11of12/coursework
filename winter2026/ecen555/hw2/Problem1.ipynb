```python
import numpy as np
import matplotlib.pyplot as plt

labels = ["Diamond", "Silicon", "Germanium"]
Eg = np.array([5, 1.1, 0.66])
n = np.array([2.4, 3.46, 4.0])
```

```python
plt.rcParams.update({'font.size': 15})

plt.plot(Eg, n, "r-")
plt.scatter(Eg, n)
plt.ylabel("n")
plt.xlabel(r"$E_g$" + r" $(eV)$")
plt.title(r"$n$ vs. $E_g$")
```

```python

x = 1/Eg
y = n**4
plt.scatter(x, y)
m, b = np.polyfit(x,y,1)
_x = np.linspace(0.1, 1.5, 1000)
plt.plot(_x, m*_x + b, "r--", alpha=0.5, label=f"Line of Best Fit: {m:.2f}"+r"$\frac{1}{E_g}$" + r" $ - $ "+ f"{np.abs(b):.2f}")


plt.ylabel(r"$n^4$")
plt.xlabel(r"1/${E_g}$" + r" $(\frac{1}{eV})$")
plt.legend()
plt.title(r"$n^4$ vs. $1/E_g$")
```
