```python
from scipy.constants import h, c, e
import numpy as np


def photocurrent(wl=800e-9, l_p=0.5e-6, L_e=0.1e-6, W=1e-6, L_h=10e-6, alpha=1e5, eta_i=1, T=1, P0=1e-6):
    f = c/wl
    return e*eta_i*T*P0/(h*f) * (np.exp(-alpha*(l_p - L_e)) - np.exp(-alpha*(l_p + W + L_h))) 

properties_p_plus_n_at_800nm = {
    "wl": 800e-9,
    "l_p": 0.5e-6,
    "L_e": 0.1e-6,
    "W": 1e-6,
    "L_h": 10e-6,
    "alpha": 1e5,
}

properties_pin_at_800nm = {
    "wl": 800e-9,
    "l_p": 0.5e-6,
    "L_e": 0.1e-6,
    "W": 30e-6,
    "L_h": 0.1e-6,
    "alpha": 1e5,
}

print(photocurrent(**properties_p_plus_n_at_800nm))
print(photocurrent(**properties_pin_at_800nm))

properties_p_plus_n_at_900nm = {
    "wl": 900e-9,
    "l_p": 0.5e-6,
    "L_e": 0.1e-6,
    "W": 1e-6,
    "L_h": 10e-6,
    "alpha": 4e4,
}

properties_pin_at_900nm = {
    "wl": 900e-9,
    "l_p": 0.5e-6,
    "L_e": 0.1e-6,
    "W": 30e-6,
    "L_h": 0.1e-6,
    "alpha": 4e4,
}
print(photocurrent(**properties_p_plus_n_at_900nm))
print(photocurrent(**properties_pin_at_900nm))

properties_p_plus_n_at_500nm = {
    "wl": 500e-9,
    "l_p": 0.5e-6,
    "L_e": 0.1e-6,
    "W": 1e-6,
    "L_h": 10e-6,
    "alpha": 0.9e6,
}

properties_pin_at_500nm = {
    "wl": 500e-9,
    "l_p": 0.5e-6,
    "L_e": 0.1e-6,
    "W": 30e-6,
    "L_h": 0.1e-6,
    "alpha": 0.9e6,
}
print(photocurrent(**properties_p_plus_n_at_500nm))
print(photocurrent(**properties_pin_at_500nm))
```
