```python
import numpy as np
from scipy.constants import speed_of_light as c

n1 = 3.00
n2 = 1.50

wl   = np.array([    15,    20,    25,    30,    40,    45,    50,    70,   100,   150,   200])*1e-6
k       = 2*np.pi/wl 
theta_0 = np.array([  77.8, 74.52,  71.5,  68.7,  63.9,  61.7, 59.74,  53.2,  46.4,  39.9, 36.45])
theta_1 = np.array([  65.2, 58.15,  51.6,  45.5,  35.5, 32.02, 30.17])

k1 = n1*k

omega = 2*np.pi*c/wl
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 4e6, 1000)

beta_0 = k1*np.sin(theta_0*np.pi/180)
beta_1 = k1[:7]*np.sin(theta_1*np.pi/180)
plt.scatter(beta_0, omega, label="TE0")
plt.scatter(beta_1, omega[:7], label="TE1")
plt.plot(x, c/n2*x, "r--", label="Slope c/n2")
plt.plot(x, c/n1*x, "b--", label="Slope c/n1")

plt.xlim(0, 1.3e6)
# plt.xlim(0, 1.8e6)
plt.ylim(0, 1.3e14)

plt.xlabel(r"$\beta _m $ ($m^{-1}$)")
plt.ylabel(r"$\omega$ (rads/s)")
plt.title(r"Dispersion Diagram $\omega$ vs. $\beta _m$")
plt.legend()
# beta_1 = k1*np.sin(theta_1*np.pi/180)
```
